```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv', ',', error_bad_lines=False)
```

```python
data[data['password'].isnull()]
```

```python
data.dropna(inplace = True)
```

```python
password_tuple = np.array(data)
```

```python
import random
random.shuffle(password_tuple)
```

```python
###y is dependent variable
y = [label[1] for label in password_tuple]
y
```

```python
x = [label[0] for label in  password_tuple]
```

```python
x
```

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'strength', data = data, palette='RdBu_r')
```

```python
def word_divide_char(inputs):
    characters = []
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
x = vectorizer.fit_transform(x)
```

```python
x.shape
```

```python
vectorizer.vocabulary_
```

```python
feature_names = vectorizer.get_feature_names()

#get tfidf vector for first document
first_document_vector = x[0]


#print the score
df = pd.DataFrame(first_document_vector.T.todense(), index = feature_names, columns = ['tfidf'])

df.sort_values(by = ['tfidf'], ascending = False)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
log_class = LogisticRegression(penalty = 'l2', multi_class='ovr')
log_class.fit(x_train, y_train)
```

```python
print(log_class.score(x_test, y_test))
```

```python
#Multinomial
clf = LogisticRegression(random_state=0, multi_class='multinomial', solver = 'newton-cg')
clf.fit(x_train, y_train)
print(clf.score(x_test, y_test))
```

```python
x_predict = np.array(['@#$'])
x_predict = vectorizer.transform(x_predict)
y_pred = log_class.predict(x_predict)
print(y_pred)
```

```python
x_predict = np.array(['dsfhbuioer263489892@#%^&'])
x_predict = vectorizer.transform(x_predict)
y_pred = clf.predict(x_predict)
print(y_pred)
```
